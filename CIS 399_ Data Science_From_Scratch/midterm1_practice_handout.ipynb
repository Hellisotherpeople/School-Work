9: Write a function to find out how many times a column is used in a tree
</h1>
<p>
<div class=h1_cell>
<p>
For a specific tree, which columns actually get used when building the tree? And how many times do they appear? Write a function `column_counts` that takes a tree in as input and produces a dictionary of columns that the tree uses as keys and how often as values. Test it on tree1 and tree2.
<p>

</div>

```python
import pprint as pprint 


def column_counts(tree):
    
    newlist = []
    for path in tree['paths']:
        conjunction = path['conjunction']
        res_list = [x[0] for x in conjunction]
        for item in res_list:
            newlist.append(item[:-2]) ## [:-2 gets rid of the _0 or _1 columns that are added by the OHE]


    counts = dict()
    for i in newlist:
        counts[i] = counts.get(i, 0) + 1
    return counts 
```

10: Find shortest path in tree
</h1>
<p>
<div class=h1_cell>
<p>
Write a function to find the shortest paths in a tree. If there is a unique shortest path then your list will just have one path. If there are ties, the list will include all that tied.
<p>
</div>

```python

# for path in tree1['paths']:
#     conjunction = path['conjunction']
#     print(len(conjunction))

def list_of_shortest(tree):
    shortest_path_list = []
    shortest_path_length = min(len(path['conjunction']) for path in tree['paths'])
    for path in tree['paths']:
        #print(path)
        if len(path['conjunction']) == shortest_path_length:
            shortest_path_list.append(path)
    return shortest_path_list
```
